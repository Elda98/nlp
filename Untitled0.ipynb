```python
import re
from typing import List


class ArabicPreprocessor:
    def preprocess(self, text: str, tokenize: bool = True):
        text = text.lower()

        text = re.sub(r'http\S+|@\S+', '', text)

        text = re.sub(r'#', '', text)

        text = re.sub(r'[\u064B-\u065F]', '', text)

        text = re.sub(r'[أإآٱ]', 'ا', text)
        text = re.sub(r'ة', 'ه', text)

        text = re.sub(r'[^\w\s]', '', text)

        text = re.sub(r'(.)\1+', r'\1', text)

        text = text.strip()

        tokens = text.split()
        return tokens if tokenize else " ".join(tokens)

processor = ArabicPreprocessor()
df = pd.read_csv('raghad.csv')

sample_tweets = df['text'].sample(5, random_state=42).tolist()

print("-" * 100)
print(f"{'Before Preprocessing (Original)':<50} | {'After Preprocessing (Cleaned)':<50}")
print("-" * 100)

for original in sample_tweets:
        cleaned = processor.preprocess(original, tokenize=False)

        print(f"{str(original)[:45]:<50} | {str(cleaned)[:45]:<50}")

```

```python
import math
import pandas as pd
from collections import Counter, defaultdict
from typing import List
from sklearn.metrics import classification_report, confusion_matrix

class NaiveBayesClassifier:
    def __init__(self, k: float = 1.0):
        self.k = k
        self.log_priors = {}
        self.word_log_likelihoods = defaultdict(lambda: defaultdict(float))
        self.vocab = set()

    def train(self, documents: List[List[str]], labels: List[str]):
        n_docs = len(labels)
        label_counts = Counter(labels)

        for label, count in label_counts.items():
            self.log_priors[label] = math.log(count / n_docs)

        word_counts_per_class = defaultdict(Counter)
        total_words_per_class = Counter()

        for doc, label in zip(documents, labels):
            for word in doc:
                word_counts_per_class[label][word] += 1
                total_words_per_class[label] += 1
                self.vocab.add(word)

        for label in label_counts:
            denom = total_words_per_class[label] + len(self.vocab)
            for word in self.vocab:
                count = word_counts_per_class[label][word]
                self.word_log_likelihoods[label][word] = math.log((count + self.k) / denom)

    def predict_one(self, tokens: List[str]) -> str:
        scores = {}
        for label in self.log_priors:
            score = self.log_priors[label]
            for word in tokens:
                if word in self.vocab:
                    score += self.word_log_likelihoods[label][word]
            scores[label] = score
        return max(scores, key=scores.get)

try:
    df = pd.read_csv('raghad.csv')
except:
    df = pd.read_csv('raghad')

df.columns = df.columns.str.strip()

text_col = 'text' if 'text' in df.columns else df.columns[0]
label_col = 'label' if 'label' in df.columns else df.columns[1]


processor = ArabicPreprocessor()
X = [processor.preprocess(str(text), tokenize=True) for text in df[text_col].tolist()]
y = df[label_col].astype(str).tolist()

split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

nb_model = NaiveBayesClassifier(k=1.0)
nb_model.train(X_train, y_train)
y_pred = [nb_model.predict_one(tokens) for tokens in X_test]

print(" Naive Bayes Final Evaluation Results")
print("\n 1. Classification Report:")
print(classification_report(y_test, y_pred))

print("\n 2. Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\n 3. Sample Predictions Table:")
print("-" * 85)
print(f"{'Cleaned Tweet (Input)':<50} | {'Pred':<6} | {'Actual':<6}")
print("-" * 85)

for i in range(min(10, len(X_test))):
    tweet_text = " ".join(X_test[i])[:48]
    print(f"{tweet_text:<50} | {y_pred[i]:<6} | {y_test[i]:<6}")

print("-" * 85)
```

```python
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
from typing import List

def evaluate_sentiment_model(file_path: str, preprocessor_obj: ArabicPreprocessor, model_obj: NaiveBayesClassifier) -> None:
    """
    Performs the final evaluation of the Naive Bayes model and prints a clean table.
    """

    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f" Error: The file '{file_path}' was not found.")
        return

    df.columns = df.columns.str.strip()
    text_col, label_col = df.columns[0], df.columns[1]

    test_sample = df.sample(min(50, len(df)), random_state=42)

    all_preds, all_actuals = [], []

    print(f"{'Actual':<10} | {'Pred':<10} | {'Cleaned Arabic Tweet'}")
    print("-" * 85)

    for _, row in test_sample.iterrows():
        tokens = preprocessor_obj.preprocess(str(row[text_col]))
        prediction = str(model_obj.predict_one(tokens))
        actual = str(row[label_col])

        all_preds.append(prediction)
        all_actuals.append(actual)

        display_text = " ".join(tokens)[:60]
        print(f"{actual:<10} | {prediction:<10} | {display_text}")


    print("\nDetailed Classification Report:")
    print(classification_report(all_actuals, all_preds))

    print("\nConfusion Matrix:")
    print(confusion_matrix(all_actuals, all_preds))

    correct = sum(1 for p, a in zip(all_preds, all_actuals) if p == a)
    final_acc = (correct / len(all_preds)) * 100
    print(f"\n Overall Project Accuracy: {final_acc:.2f}%")

evaluate_sentiment_model('raghad.csv', preprocessor, nb_model)
```
